```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, butter, filtfilt, detrend

# Load the binary .dat file
data = np.fromfile("../data/data_bin/40689238.dat", dtype=np.int16)

# Define number of leads (usually 12 for standard ECG)
num_leads = 12

# Reshape into (samples, leads)
num_samples = len(data) // num_leads
data = data.reshape((num_samples, num_leads))

# Apply scaling factors (Assume these are provided)
scale_factors = np.array([200.0] * num_leads)  # Example scale factors
data = data / scale_factors  # Convert to mV


# Create separate plots for each lead
for i in range(num_leads):
    plt.figure(figsize=(16, 4))
    plt.plot(data[:, i], label=f'Lead {i+1}')
    plt.title(f"ECG Lead {i+1}")
    plt.xlabel("Sample")
    plt.ylabel("mV")
    plt.legend()
    plt.grid(True)
    plt.show()

plt.figure(figsize=(12, 6))
for i in range(num_leads):
    plt.plot(data[:, i] + i * 2, label=f'Lead {i+1}')  # Offset for visibility

plt.legend()
plt.title("ECG - All 12 Leads")
plt.xlabel("Sample")
plt.ylabel("mV (with offset)")
plt.show()
```
